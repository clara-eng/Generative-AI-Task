## 🔧 Phase 1: Environment & Model Setup (10 min)

We'll use GPT-2 from Hugging Face to observe how prompting strategies affect output quality.
GPT-2 is a good teaching model because it clearly shows the difference between standard and CoT prompting.

```python
# Install required packages
!pip install -q transformers accelerate torch
```

```python
import warnings
warnings.filterwarnings('ignore')

from transformers import pipeline, set_seed
import textwrap

# Load text generation model
print("Loading GPT-2...")
generator = pipeline('text-generation', model='gpt2', device=-1)
print("✅ Model loaded!")

def generate(prompt, max_length=200, seed=42):
    """Generate a single completion for a prompt."""
    set_seed(seed)
    result = generator(
        prompt,
        max_length=max_length,
        num_return_sequences=1,
        do_sample=True,
        temperature=0.7,
        top_p=0.9,
        truncation=True
    )
    return result[0]['generated_text']

def compare_prompts(prompt_standard, prompt_cot, max_length=200):
    """Side-by-side comparison of standard vs CoT prompting."""
    print("=" * 70)
    print("🔴 STANDARD PROMPT:")
    print(f"   {prompt_standard}")
    output_std = generate(prompt_standard, max_length=max_length)
    print(f"\n📤 Output:\n{textwrap.fill(output_std, width=70)}")

    print("\n" + "-" * 70)
    print("🟢 COT PROMPT:")
    print(f"   {prompt_cot}")
    output_cot = generate(prompt_cot, max_length=max_length)
    print(f"\n📤 Output:\n{textwrap.fill(output_cot, width=70)}")
    print("=" * 70)
    return output_std, output_cot

# Quick test
test = generate("The capital of France is", max_length=30)
print(f"\n🧪 Test: {test}")
```

## 🧠 Phase 2: Zero-Shot CoT — The Power of "Let's think step by step" (10 min)

### Concept Recap
Zero-Shot CoT uses a **single trigger phrase** to activate reasoning without providing examples.
Common triggers: "Let's think step by step", "Explain your reasoning", "Break down the solution".

### Exercise 2.1: Math Problem — Standard vs Zero-Shot CoT

```python
# Let's see how adding "Let's think step by step" changes the output

# Standard prompt (no CoT)
prompt_standard = "Q: A store sells a product for $80 after a 20% discount. What was the original price?\nA:"

# Zero-Shot CoT prompt
prompt_zscot = "Q: A store sells a product for $80 after a 20% discount. What was the original price?\nA: Let's think step by step."

compare_prompts(prompt_standard, prompt_zscot, max_length=150)
```

### Exercise 2.2: Your Turn — Zero-Shot CoT Trigger Phrases

Try **three different trigger phrases** on the same problem and compare which gives the best reasoning.

```python
# The problem to solve
problem = "Q: A train travels 120 km in 2 hours. Then it speeds up and travels 180 km in 2 hours. What is the average speed for the entire trip?\nA:"

trigger_phrases = [
    "Let's think step by step.",
    "Let me break this down.",
    "First, let me identify the key information."
]

for i, trigger in enumerate(trigger_phrases):
    prompt = problem.rstrip() + " " + trigger
    output = generate(prompt, max_length=200)
    print(f"\n{'='*60}")
    print(f"Trigger {i+1}: {trigger}")
    print(f"Output: {output}")
```

## 📚 Phase 3: Few-Shot CoT — Teaching by Example (15 min)

### Concept Recap
Few-Shot CoT provides **worked examples with reasoning steps** before asking the actual question.
The model learns the *pattern* of reasoning from the examples.

### Exercise 3.1: Few-Shot vs Standard Few-Shot

```python
# Standard Few-Shot (answer only, no reasoning)
prompt_fewshot = """Q: Roger has 5 tennis balls. He buys 2 cans of 3 balls each. How many does he have?
A: The answer is 11.

Q: A bakery makes 40 loaves in the morning and 30 in the afternoon. They sell 50. How many are left?
A:"""

# Few-Shot CoT (reasoning shown in examples)
prompt_fewshot_cot = """Q: Roger has 5 tennis balls. He buys 2 cans of 3 balls each. How many does he have?
A: Roger starts with 5 balls. He buys 2 cans, each containing 3 balls. So he buys 2 * 3 = 6 balls. Total = 5 + 6 = 11. The answer is 11.

Q: A bakery makes 40 loaves in the morning and 30 in the afternoon. They sell 50. How many are left?
A:"""

compare_prompts(prompt_fewshot, prompt_fewshot_cot, max_length=200)
```

### Exercise 3.2: Build Your Own Few-Shot CoT Prompt

Create a Few-Shot CoT prompt for **logical reasoning** (not math). Provide 2 worked examples, then ask a new question.

```python
fewshot_cot_prompt = """Q: The road is wet and slippery after heavy rain. Is it safe to drive at high speed?
A: Let's think step by step. Wet roads reduce tire friction. Reduced friction increases stopping distance. High speed requires more stopping distance. Therefore, driving at high speed on a wet road is NOT safe. Answer: No.

Q: Sara studied hard every night for a week before her exam. She also slept 8 hours the night before. Is she likely to perform well?
A: Let's think step by step. Consistent studying builds knowledge. 8 hours of sleep improves memory consolidation and focus. Both factors contribute positively to exam performance. Therefore, Sara is likely to perform well. Answer: Yes.

Q: Ahmed skipped breakfast, stayed up all night, and has an important job interview in the morning. Will he likely perform at his best?
A:"""

output = generate(fewshot_cot_prompt, max_length=250)
print(output)
```

## 🔨 Phase 4: CoT Templates in Practice (15 min)

### Concept Recap
The slides introduced several CoT templates: General, Analytical, Mathematical, Coding & Debugging.
Each template structures reasoning differently depending on the domain.

### 📝 Practice Exercise 1: Mathematical CoT Template
**From Slide 23:** *"A store sells a product for $80 after a 20% discount. What was the original price?"*

Apply the **Mathematical CoT Template**:
1. Identify knowns
2. Identify unknowns
3. Convert to equation
4. Solve symbolically
5. Check units & constraints

```python
math_cot_prompt = """Solve the following problem using the Mathematical CoT Template:
Step 1 - Identify knowns: List all given values.
Step 2 - Identify unknowns: State what we need to find.
Step 3 - Convert to equation: Write the algebraic relationship.
Step 4 - Solve symbolically: Show each algebraic step.
Step 5 - Check units & constraints: Verify the answer makes sense.

Problem: A store sells a product for $80 after a 20% discount. What was the original price?

Step 1 - Knowns:"""

output = generate(math_cot_prompt, max_length=250)
print(output)
```

### 📝 Practice Exercise 2: Analytical CoT Template (Structured CoT)
**From Slide 24:** *"Explain why global supply chains became unstable in 2021"*

Apply the **Analytical CoT Template**:
1. Describe the situation
2. Identify constraints
3. Evaluate options/factors
4. Justify analysis
5. Deliver final conclusion

```python
analytical_cot_prompt = """Analyze the following question using the Analytical CoT Template:
Step 1 - Situation: Describe the context and background.
Step 2 - Constraints: What limitations or pressures existed?
Step 3 - Factors: Evaluate the key contributing causes.
Step 4 - Justify: Explain which factors were most significant and why.
Step 5 - Conclusion: Deliver a final, well-reasoned answer.

Question: Explain why global supply chains became unstable in 2021.

Step 1 - Situation:"""

output = generate(analytical_cot_prompt, max_length=300)
print(output)
```

### 📝 Practice Exercise 3: Build Reasoning Prompts for 3 Domains
**From Slide 25:** Design CoT prompts for:
1. A **logic puzzle**
2. A **coding error**
3. A **business scenario**

Use the appropriate CoT template for each!

```python
logic_puzzle_prompt = """Solve this logic puzzle using the General CoT Framework:
Step 1 - Restate: Summarize the puzzle in your own words.
Step 2 - Identify: List all clues and constraints.
Step 3 - Break down: Work through each clue one by one.
Step 4 - Solve: Combine deductions to find the answer.
Step 5 - Verify: Check the solution against all clues.

Puzzle: Three friends — Ali, Sara, and Omar — each have a different pet: a cat, a dog, and a fish.
Clue 1: Ali does not have the cat.
Clue 2: Sara does not have the dog.
Clue 3: Omar does not have the fish.
Who has which pet?

Step 1 - Restate:"""

output = generate(logic_puzzle_prompt, max_length=300)
print(output)
```

```python
coding_error_prompt = """Debug the following Python code using the Coding & Debugging CoT Template:
Step 1 - Understand: What is the code supposed to do?
Step 2 - Locate: Where does the error occur?
Step 3 - Explain: Why is this line wrong?
Step 4 - Fix: What is the correct code?
Step 5 - Corrected code: Show the full corrected version.

Buggy code:
def calculate_average(numbers):
    total = 0
    for num in numbers:
        total += num
    return total / len(numbers)

result = calculate_average([])
print(result)

Step 1 - Understand:"""

output = generate(coding_error_prompt, max_length=300)
print(output)
```

```python
business_prompt = """Analyze this business decision using the Analytical CoT Template:
Step 1 - Situation: Describe the business context.
Step 2 - Constraints: What are the budget and resource limits?
Step 3 - Options: What are the possible choices?
Step 4 - Justify: Weigh the pros and cons of each option.
Step 5 - Recommend: Give a clear final recommendation.

Scenario: A startup has a $100K budget and must choose between hiring a new developer or investing in a marketing campaign. They have a working product but low user growth.

Step 1 - Situation:"""

output = generate(business_prompt, max_length=300)
print(output)
```

## 🧪 Phase 5: Role-Enhanced CoT & Prompt Decomposition (10 min)

### Concept Recap
- **Role-Enhanced CoT**: Combine a role ("You are a senior mathematician") with CoT instructions
- **Prompt Decomposition**: Break one complex prompt into multiple simpler prompts, chain results

### Exercise 5.1: Role-Enhanced CoT

```python
prompt_no_role = """Explain why a neural network might overfit on a small dataset.
Let's think step by step:
Step 1 - Define overfitting:
Step 2 - Explain how dataset size relates to model complexity:
Step 3 - Describe what happens during training on small data:
Step 4 - Conclusion:"""

prompt_with_role = """You are a senior ML engineer with 10 years of experience teaching deep learning.
Explain to a junior developer why a neural network might overfit on a small dataset.
Think step by step:
Step 1 - Define overfitting:
Step 2 - Explain how dataset size relates to model complexity:
Step 3 - Describe what happens during training on small data:
Step 4 - Practical advice to prevent it:"""

compare_prompts(prompt_no_role, prompt_with_role)
```

### Exercise 5.2: Prompt Decomposition

Break a complex question into a chain of smaller prompts. Feed the output of each into the next.

```python
# Decomposed prompt chain: ROI of AI for a mid-size Egyptian company

# Step 1: Identify factors that affect ROI
prompt_1 = """What are the main factors that affect the ROI of investing in AI for a business?
List them step by step with a brief explanation of each.
Factor 1:"""
output_1 = generate(prompt_1, max_length=150)
print("=" * 60)
print("STEP 1 — Key ROI Factors:")
print(output_1)

# Step 2: Estimate costs (feed context from step 1)
prompt_2 = f"{output_1}\n\nBased on the above factors, estimate the typical costs a mid-size company would face when implementing AI. Include setup, talent, and maintenance costs.\nCost 1:"
output_2 = generate(prompt_2, max_length=150)
print("\n" + "=" * 60)
print("STEP 2 — Estimated Costs:")
print(output_2)

# Step 3: Estimate benefits/savings
prompt_3 = f"{output_2}\n\nNow estimate the typical benefits and savings a mid-size company could gain from AI adoption (efficiency, revenue, cost savings).\nBenefit 1:"
output_3 = generate(prompt_3, max_length=150)
print("\n" + "=" * 60)
print("STEP 3 — Estimated Benefits:")
print(output_3)

# Step 4: Final ROI assessment
prompt_4 = f"{output_3}\n\nBased on the costs and benefits above, provide a final ROI assessment for a mid-size Egyptian company investing in AI. Should they invest? Justify your answer.\nFinal Assessment:"
output_4 = generate(prompt_4, max_length=200)
print("\n" + "=" * 60)
print("STEP 4 — Final ROI Assessment:")
print(output_4)
```
